```python
import requests
from google.colab import userdata
```

```python
baseurl = 'https://q.utoronto.ca/api/v1/'

access_token = userdata.get('q_access_token')

headers = {
    "Authorization": f"Bearer {access_token}"
}
```

### GET Courses

```python
def get_all_courses():
    courses = []
    page = 1

    while True:
        url = f"{baseurl}courses?include%5B%5D=enrollments&per_page=100&page={page}"

        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            data = response.json()

            courses.extend(data)

            if 'next' in response.links:
                page += 1
            else:
                break
        else:
            print(f"Error fetching data: {response.status_code}")
            break

    return courses

all_courses = get_all_courses()

# Print out the details for each course
for course in all_courses:
    print(course)
```

### GET Assignments

```python
def get_assignments(course_id):
    assignments = []
    page = 1

    while True:
        # Define the URL for the current page with the appropriate parameters
        url = f"{baseurl}courses/{course_id}/assignments?per_page=100&page={page}"

        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            data = response.json()  # Parse the JSON response

            assignments.extend(data)

            if 'next' in response.links:
                page += 1
            else:
                break
        else:
            print(f"Error fetching assignments: {response.status_code}")
            break

    return assignments


def get_submission_for_assignment(course_id, assignment_id):
    url = f"{baseurl}courses/{course_id}/assignments/{assignment_id}/submissions"
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error - {assignment_id}: {response.status_code}")
        return None


assignments = get_assignments(357793)


for assignment in assignments:
    print(f"Assignment: {assignment['name']}")
    print(f"Due Date: {assignment['due_at']}")
    print(f"Lock Date (Until Date): {assignment['lock_at']}")

    submission_data = get_submission_for_assignment(assignment['course_id'], assignment['id'])

    if submission_data:
        submission_time = submission_data.get('submitted_at', 'Not submitted yet')
        print(f"Submitted At: {submission_time}")
    else:
        print("No submission data available.")

    print("-" * 50)
```
